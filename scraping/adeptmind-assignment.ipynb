## scrapingclub.com

### https://scrapingclub.com/exercise/detail_basic/

```python
product_data = dict()
url = 'https://scrapingclub.com/exercise/detail_basic/'
#url = 'https://scrapingclub.com/exercise/detail_json/'    # information for this url is coming in a json object
# url = 'https://scrapingclub.com/exercise/detail_ajax/'    # Need to fetch information using Ajax call 
#url = 'https://scrapingclub.com/exercise/list_detail_infinite_scroll/93086-B/' # Need to fetch information using Ajax call 
#url = 'https://scrapingclub.com/exercise/detail_header/'  #Ajax
#url = 'https://scrapingclub.com/exercise/detail_cookie/'
#url = 'https://scrapingclub.com/exercise/detail_sign/'

base_url = urlparse(url).netloc
print(base_url)

page = requests.get(url)

#print(page.content)

soup = BeautifulSoup(page.content, "html.parser")

product_data['title'] = soup.find("h3", class_="card-title").text
product_data['price'] = soup.find("div", class_="card-body").find("h4").text
product_data['image'] = 'https://scrapingclub.com/exercise/detail_basic' + soup.find("img", class_="card-img-top img-fluid")['src']
product_data['description'] = soup.find("div", class_="card-body").find("p").text

product_data
```

### 'https://scrapingclub.com/exercise/detail_json/' 

```python
url = 'https://scrapingclub.com/exercise/detail_json/'    # information for this url is coming in a json object
base_url = urlparse(url).netloc
page = requests.get(url)
soup=BeautifulSoup(page.content,"lxml")

#get to what we want
script=soup.find("script",string=re.compile('var obj ='))
json_text=re.search(r'^\s*var obj\s*=\s*({.*?})\s*;\s*$',script.string, flags=re.DOTALL | re.MULTILINE).group(1)

json_text
```

```python
json.loads(json_text.replace('" + "', ''))
#parsed=json.loads(json_text)
```

```python
#validating json
json_text=json_text.replace("\"/static/img/\" + ","")
json_text=json_text.replace("\"96230-C\" + ","")

parsed=json.loads(json_text)
print("Item Name: {} \nItem Price: {}\nItem Description: {}".format(parsed['title'],parsed['price'],parsed['description']))
```
